```python
import pandas as pd
import numpy as np
```

### Function to fill in random blanks

```python
# If there is a null value, then take in the previous value
# After data starts, then if it hits a null value, it will replace the null value with the previous value
def check_for_null(zipcode:list) -> list:
    new_list= []
    counter = 0
    for x in range(len(zipcode)):
        if not counter:
            new_list.append(np.nan)
            if zipcode[x]:
                counter =1 
        else: 
            if not zipcode[x] and not zipcode[x-1]:
                new_list.append(zipcode[x-2])
            else: 
                new_list.append(zipcode[x]) if zipcode[x] else new_list.append(zipcode[x-1])

    return new_list
```

```python
#Check to ensure function works as designed 
zip = [0,0,0,6,8,4,8,0,3,5,4]
check_for_null(zip)
```

### Function to calculate percentage change

```python
# Put 0 as a placeholder for the first date. 
# Calculate the percentage change in the rate since the previous date 
def percent_change(zipcode:list) -> list:
    new_list =[0,]
    for x in range(1, len(zipcode)): 
        new_list.append(round((100*(float(zipcode[x])-float(zipcode[x-1]))/float(zipcode[x-1])), 2))
    return new_list
```

# Zestimates

```python
df_zestimates = pd.read_csv("../Resources/zestimates_over_time.csv")
df_zestimates
```

```python
# Rename the Date column
df_zestimates.rename(columns ={"Unnamed: 0": "Date"}, inplace= True)

# Grab the date column and store it
dates = df_zestimates["Date"]
```

```python
# Get a list of the columns for the zestimates
columns_zest = df_zestimates.columns
```

```python
zestimate_changes = {
    x: percent_change(df_zestimates[x]) for x in columns_zest[1:]
}
```

```python
df_zestimate_changes = pd.DataFrame(zestimate_changes)
df_zestimate_changes.set_index(dates, inplace = True)
```

```python
df_zestimate_changes.reset_index(inplace=True)
```

```python
df_zestimate_changes
```

```python
df_zestimate_changes.to_csv("../Resources/zestimate_changes_over_time.csv", index=False)
```
