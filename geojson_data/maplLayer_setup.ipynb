```python
import pandas as pd
import json
```

```python
#reading json
geojson_df = pd.read_json('../geojson_data/tx_texas_zip_codes_geo.min.json')
```

```python
# load data using Python JSON module
with open('../geojson_data/tx_texas_zip_codes_geo.min.json','r') as f:
    data = json.loads(f.read())
```

```python
# Normalizing data
geo_df = pd.json_normalize(data, record_path = 
                           ['features'], meta =
                           ['original_number_of_clusters','Scaler','family_min_samples_percentage'], meta_prefix='config_params_', record_prefix='dbscan_', errors = 'ignore')
```

```python
geo_df.head(3)
```

```python
#Getting specific data points for the Austin area
geo_with_index = geo_df.set_index("dbscan_properties.ZCTA5CE10")
geo_with_index.head(3)
```

```python
zipcodes = ['78744',
            '78735',
            '78741',
            '78733',
            '78746',
            '78701',
            '78702',
            '78721',
            '78722',
            '78723',
            '78754',
            '78731',
            '78732',
            '78726',
            '78750',
            '78757',
            '78752',
            '78724',
            '78753',
            '78758',
            '78727',
            '78759',
            '78745',
            '78729',
            '78748',
            '78717',
            '78751',
            '78705',
            '78704',
            '78703']
austingeo_df = geo_df[geo_df['dbscan_properties.ZCTA5CE10'].isin(zipcodes)]
```

```python
austingeo_df.head()
```

```python
austin_df = austingeo_df.drop(["config_params_original_number_of_clusters", "config_params_Scaler", 'config_params_family_min_samples_percentage'], axis =1)
```

```python
austin_df.head()
```

```python
# Renaming Columns for geojson
austin_df.rename(columns= {'dbscan_type':'type',
                        'dbscan_properties.STATEFP10' : 'STATEFP10',
                       'dbscan_properties.ZCTA5CE10' : 'ZCTA5CE10',
                       'dbscan_properties.GEOID10' : 'GEOID10',
                       'dbscan_properties.CLASSFP10' : 'CLASSFP10',
                       'dbscan_properties.MTFCC10' : 'MTFCC10',
                       'dbscan_properties.FUNCSTAT10' : 'FUNCSTAT10',
                       'dbscan_properties.ALAND10' : 'ALAND10',
                       'dbscan_properties.AWATER10': 'AWATER10',
                       'dbscan_properties.INTPTLAT10': 'INTPTLAT10',
                       'dbscan_properties.INTPTLON10': 'INTPTLON10',
                       'dbscan_properties.PARTFLG10': 'PARTFLG10',
                       'dbscan_geometry.type': 'geometry.type',
                       'dbscan_geometry.coordinates':'geometry.coordinates'},
             inplace=True)
```

```python
austin_df.dtypes
```

```python
austin_df.to_json('..//geojson_data/austinzipcode_geo.json', orient="records")
austin_df.to_csv("../geojson_data/austinzipcode_geo.csv", index=False)
```

```python
rent_df = pd.read_csv('../geojson_data/rentals_over_timeGeo.csv')
zest_df = pd.read_csv('../geojson_data/zest_over_timeGeo.csv')
```

```python
rent_df.dtypes
```

```python
austin_merge = austin_df.rename(columns= {'ZCTA5CE10' : 'zip_code'})
austin_merge = austin_merge.astype({"zip_code": int})
```

```python
austin_merge = austin_merge.drop(["type","GEOID10", "CLASSFP10", "MTFCC10", "FUNCSTAT10", "ALAND10", "AWATER10", "STATEFP10", "PARTFLG10", "geometry.type", "geometry.coordinates"], axis=1)
```

```python
rent_geo = pd.merge(rent_df, austin_merge, on="zip_code", how="left")
zest_geo = pd.merge(zest_df, austin_merge, on="zip_code", how="left")
rent_geo = rent_geo.rename(columns= {'zip_code' : 'ZCTA5CE10',
                           'INTPTLON10': 'longitude',
                           'INTPTLAT10': 'latitude'})
zest_geo = zest_geo.rename(columns= {'zip_code' : 'ZCTA5CE10',
                           'INTPTLON10': 'longitude',
                           'INTPTLAT10': 'latitude'})
```

```python
#dropping na values in lan and long
rent_geo = rent_geo.dropna(subset=['latitude', 'longitude'])
zest_geo = zest_geo.dropna(subset=['latitude', 'longitude'])
```

```python
rent_geo.head()
```

```python
rent_geo.to_csv('../geojson_data/rent_geo.csv', index=False)
zest_geo.to_csv('../geojson_data/zest_geo.csv', index=False)

rent_geo.to_json('..//geojson_data/geo_jsonmap/rent_geo.json', orient="records")
zest_geo.to_json('..//geojson_data/geo_jsonmap/zest_geo.json', orient="records")
```
