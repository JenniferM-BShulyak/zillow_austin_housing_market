```python
import pandas as pd
import json
```

```python
#reading json
geojson_df = pd.read_json('../geojson_data/tx_texas_zip_codes_geo.min.json')
```

```python
geojson_df.head(3)
```

```python
# load data using Python JSON module
with open('../geojson_data/tx_texas_zip_codes_geo.min.json','r') as f:
    data = json.loads(f.read())
```

```python
# Normalizing data
geo_df = pd.json_normalize(data, record_path = 
                           ['features'], meta =
                           ['original_number_of_clusters','Scaler','family_min_samples_percentage'], meta_prefix='config_params_', record_prefix='dbscan_', errors = 'ignore')
```

```python
geo_df.head(3)
```

```python
#Getting specific data points for the Austin area
geo_with_index = geo_df.set_index("dbscan_properties.ZCTA5CE10")
geo_with_index.head(3)
```

```python
zipcodes = ['78744',
            '78735',
            '78741',
            '78733',
            '78746',
            '78701',
            '78702',
            '78721',
            '78722',
            '78723',
            '78754',
            '78731',
            '78732',
            '78726',
            '78750',
            '78757',
            '78752',
            '78724',
            '78753',
            '78758',
            '78727',
            '78759',
            '78745',
            '78729',
            '78748',
            '78717',
            '78751',
            '78705',
            '78704',
            '78703']
austingeo_df = geo_df[geo_df['dbscan_properties.ZCTA5CE10'].isin(zipcodes)]
```

```python
austingeo_df.head()
```

```python
austin_df = austingeo_df.drop(["config_params_original_number_of_clusters", "config_params_Scaler", 'config_params_family_min_samples_percentage'], axis =1)
```

```python
austin_df.head()
```

```python
austin_df.rename(columns= {'dbscan_type':'type',
                        'dbscan_properties.STATEFP10' : 'properties.STATEFP10',
                       'dbscan_properties.ZCTA5CE10' : 'properties.Zip_code',
                       'dbscan_properties.GEOID10' : 'properties.GEOID10',
                       'dbscan_properties.CLASSFP10' : 'properties.CLASSFP10',
                       'dbscan_properties.MTFCC10' : 'properties.MTFCC10',
                       'dbscan_properties.FUNCSTAT10' : 'properties.FUNCSTAT10',
                       'dbscan_properties.ALAND10' : 'properties.ALAND10',
                       'dbscan_properties.AWATER10': 'properties.AWATER10',
                       'dbscan_properties.INTPTLAT10': 'properties.INTPTLAT10',
                       'dbscan_properties.PARTFLG10': 'properties.PARTFLG10',
                       'dbscan_geometry.type': 'geometry.type',
                       'dbscan_geometry.coordinates':'geometry.coordinates'},
             inplace=True)
```

```python
austin_df.to_file('../geojson_data/ausinzipcode_geo.geojson', driver='GeoJSON')  
#austin_df.to_csv("../geojson_data/ausinzipcode_geo.csv", index=False)
```
